```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
# Data Generation, d=2
X2, y2 = make_regression(n_samples=500, n_features=20, n_informative= 15, bias= 1, noise= 20, n_targets=2, random_state=55)
np.random.seed(650)
np.transpose(y2)[:1]+= np.random.randn(500)*50
df2 = pd.DataFrame(X2)
df2[['r1', 'r2']] = y2
df2.to_csv("Data2Dnoised500.csv")
```
