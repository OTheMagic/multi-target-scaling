# Gaussian: variaous noise

```python
## Import data
bonferoni0 = pd.read_csv("syn_exps/gaussian/bonferroni_gaussian.csv")
scaled_shortcut0 = pd.read_csv("syn_exps/gaussian/scaled_shortcut_gaussian.csv")
standardized_shortcut0 = pd.read_csv("syn_exps/gaussian/standardized_shortcut_gaussian.csv")
scaled_baseline0 = pd.read_csv("syn_exps/gaussian/scaled_baseline_gaussian.csv")
standardized_baseline0 = pd.read_csv("syn_exps/gaussian/standardized_baseline_gaussian.csv")
unscaled0 = pd.read_csv("syn_exps/gaussian/unscaled_gaussian.csv")
point_chr0 = pd.read_csv("syn_exps/gaussian/point_chr_gaussian.csv")
empirical_copula0 = pd.read_csv("syn_exps/gaussian/ecop_gaussian.csv")

df_dict0 = {
    "Unscaled": unscaled0,
    "Bonferroni": bonferoni0,
    "Point CHR": point_chr0,
    "Scaled-Splitting": scaled_baseline0,
    "Standardized-Splitting": standardized_baseline0,
    "Empirical Copula": empirical_copula0,
    "Scaled-Shortcut": scaled_shortcut0,
    "Standardized-Shortcut": standardized_shortcut0
}
```
